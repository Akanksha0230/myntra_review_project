```python
import os 
import time
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
product = "anouk kurta"
# Define the path to chrome WEbdriver executable

#Initialize the chrome webdriver 
driver= webdriver.Chrome()
# Navigate to url
driver.get("https://www.myntra.com/kurta?rawQuery="+product)

#once the page is loaded, you can extract the page text
myntra_text= driver.page_source
myntra_html = bs(myntra_text,"html.parser")
```

```python
myntra_text
```

```python
myntra_html
```

```python
pclass = myntra_html.find_all("ul",{"class":"results-base"})
```

```python
pclass[0]
```

```python
for i in pclass:
    href = i.find_all('a',href=True)
    t = (href[0]['href'])
productLink= "https://www.myntra.com/"+t
productLink
```

```python
driver.get(productLink)
proRes= driver.page_source
```

```python
proRes_html= bs(proRes,"html.parser")
proRes_html
```

```python
title_h= proRes_html.find_all("title")
title_h
```

```python
title= title_h[0].text
title
```

```python
overallRating = proRes_html.find_all("div",{"class":"index-overallRating"})
overallRating
```

```python
for i in overallRating:
    rating_value = i.find('div').text
print(rating_value)
```

```python
price= proRes_html.find_all("span",{"class":"pdp-price"})
price
```
